# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine
import time
```

# Generate data

```python
N = 1000
dimensions = list(range(10, 101, 50)) + list(range(200, 501, 100))
```

# Execute each number of dimensions

```python
standard_deviations = []
execution_times = []

# Execution for each number of dimensions
for d in dimensions:
    # Generate random points
    points = np.random.rand(N, d)
    
    # Select a random point as reference
    reference_point = points[np.random.choice(N)]
    
    # Measure execution time
    start_time = time.time()
    
    # Calculate cosine similarity between reference point and all other points
    distances = [cosine(reference_point, points[i]) for i in range(N)]
    
    # Calculate standard deviation
    std_dev = np.std(distances)
    standard_deviations.append(std_dev)
    
    # Record execution time
    total_time = time.time() - start_time
    execution_times.append(total_time)
```

# Plot the standard deviation of distances

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(dimensions, standard_deviations, marker='o', color='b')
plt.title('Standard Deviation of Distances vs. Dimensions')
plt.xlabel('Dimensions')
plt.ylabel('Standard Deviation')

# Plot the execution time
plt.subplot(1, 2, 2)
plt.plot(dimensions, execution_times, marker='o', color='r')
plt.title('Execution Time vs. Dimensions')
plt.xlabel('Dimensions')
plt.ylabel('Execution Time (seconds)')

plt.tight_layout()
plt.show()
```

# Conclusions
Desviacin Estndar: Podemos observar que la desviacin estndar de las distancias de coseno disminuye a medida que aumentan las dimensiones.
